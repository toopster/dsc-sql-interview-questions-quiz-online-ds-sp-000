## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

**One-to-one (1:1)**
One record in the primary (key) table relates to a single matching record in the foreign (key) table

**One-to-many (1:N)**
One record in the primary (key) table relates to one or more matching records in the foreign (key) table

**Many-to-many (N:M)**
One or more records in the primary (key) table relate to one or more matching records in the foreign (key) table


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

**INNER JOIN:** Returns records that have matching values in both tables

**LEFT JOIN:** Returns all records from the left / primary table, and the matched records from the right / foreign table

**RIGHT JOIN:** Returns all records from the right / foreign table, and the matched records from the left / primary table

**FULL OUTER JOIN**: Returns all records when there is a match in either left / primary or right / foreign table


## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Aggregate functions are typically used in conjunction with a GROUP BY statement.  The GROUP BY statement produces a summary record for a field with the Aggregate function defining the basis on which that summary is produced, e.g. SUM, AVG, MAX, MIN etc.


## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?


Type your answer below this line:
_______________________________________________________________________________________________________________________________

An 'Associative Entity' is a mapping table, that creates relationships between two other database tables by solely referencing the primary keys of each table. In effect, it contains a number of foreign keys, each in a many-to-one relationship from the mapping table (associative entity) to the individual data tables.
